```python
#import necessary libraries
import pandas as pd
import json
```

#Preparing Match Summery

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
with open('/content/drive/MyDrive/t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

match_df = pd.DataFrame(data[0]['matchSummary'])
match_df.head()
```

```python
#shape of the data frame
match_df.shape
```

```python
match_df.rename({"scorecard" : "match_id"}, axis = 1, inplace = True)
match_df.head()
```

```python
match_df
```

```python
match_df.duplicated().sum()
```

```python
match_df.isna().sum()
```

```python
match_ids_dict = {}

for index, row in match_df.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
match_df.to_csv('dim_match_summary.csv', index = False)
```

#Batting Summary


```python
with open('/content/drive/MyDrive/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
  
batting_df = pd.DataFrame(all_records)
batting_df.head(11)
```

```python
batting_df['out/not_out'] = batting_df.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
batting_df.head(11)
```

```python
batting_df['match_id'] = batting_df['match'].map(match_ids_dict)
batting_df.head()
```

```python
batting_df.drop(columns=["dismissal"], inplace=True)
batting_df.head(10)
```

```python
batting_df['batsmanName'] = batting_df['batsmanName'].apply(lambda x: x.replace('', ''))
batting_df['batsmanName'] = batting_df['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
batting_df.head()
```

```python
batting_df.shape
```

```python
batting_df.to_csv('fact_bating_summary.csv', index = False)
```

#Bowling Summary Preprocessing
